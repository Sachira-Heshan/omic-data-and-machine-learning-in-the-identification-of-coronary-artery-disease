```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Research/Datasets/GSE109068_Matrix_processed.csv')
```

```python
df.loc['target_value'] = (df.columns.str.startswith('HP').astype(int))
df.loc['target_value'] = df.loc['target_value'].astype(int)
```

```python
num_rows = len(df)
rows_per_set = num_rows // 15

# Split the DataFrame into fifteen sets
set1 = df.iloc[:rows_per_set, :]
set2 = df.iloc[rows_per_set:2 * rows_per_set, :]
set3 = df.iloc[2 * rows_per_set:3 * rows_per_set, :]
set4 = df.iloc[3 * rows_per_set:4 * rows_per_set, :]
set5 = df.iloc[4 * rows_per_set:5 * rows_per_set, :]
set6 = df.iloc[5 * rows_per_set:6 * rows_per_set, :]
set7 = df.iloc[6 * rows_per_set:7 * rows_per_set, :]
set8 = df.iloc[7 * rows_per_set:8 * rows_per_set, :]
set9 = df.iloc[8 * rows_per_set:9 * rows_per_set, :]
set10 = df.iloc[9 * rows_per_set:10 * rows_per_set, :]
set11 = df.iloc[10 * rows_per_set:11 * rows_per_set, :]
set12 = df.iloc[11 * rows_per_set:12 * rows_per_set, :]
set13 = df.iloc[12 * rows_per_set:13 * rows_per_set, :]
set14 = df.iloc[13 * rows_per_set:14 * rows_per_set, :]
set15 = df.iloc[14 * rows_per_set:, :]

target_value_row = df.loc['target_value']

set1.loc['target_value'] = target_value_row
set2.loc['target_value'] = target_value_row
set3.loc['target_value'] = target_value_row
set4.loc['target_value'] = target_value_row
set5.loc['target_value'] = target_value_row
set6.loc['target_value'] = target_value_row
set7.loc['target_value'] = target_value_row
set8.loc['target_value'] = target_value_row
set9.loc['target_value'] = target_value_row
set10.loc['target_value'] = target_value_row
set11.loc['target_value'] = target_value_row
set12.loc['target_value'] = target_value_row
set13.loc['target_value'] = target_value_row
set14.loc['target_value'] = target_value_row
set15.loc['target_value'] = target_value_row

print("Set 1 shape:", set1.shape)
```

```python
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest

datasets = [set1, set2, set3, set4, set5, set6, set7, set8, set9, set10, set11, set12, set13, set14, set15]

for i, dataset in enumerate(datasets, start=1):
    dataset = dataset.T
    dataset['target_value'] = dataset['target_value'].astype(int)
    score1_df = dataset[dataset.index.str.endswith('Score')]
    column_indexes1 = dataset.loc['Name'].tolist()
    score1_df.columns = column_indexes1
    score1_df.columns = list(score1_df.columns[:-1]) + ['target_value']
    score1_df.value_counts("target_value")
    X1 = score1_df.drop(columns=['target_value'])
    y1 = score1_df['target_value']

    score1_df.isnull().sum().sum()
    score1_df = score1_df.apply(lambda col: col.fillna(col.mean()))
    score1_df.isnull().sum().sum()

    # Mutual Information
    mi_classif = SelectKBest(score_func=mutual_info_classif, k=500)
    X_MI_selected = mi_classif.fit_transform(X1, y1)
    selected_features = mi_classif.get_support()
    selected_feature_names = X1.columns[selected_features]
    X1_new = X1[selected_feature_names]
    print("Shape of X1_new: ", X1_new.shape)

    X1_dataset = X1_new.copy()
    X1_dataset['target_value'] = y1

    # Calculate Mutual Information between each feature and the target variable
    mi_scores1 = mutual_info_classif(X1, y1, discrete_features=[False]*X1.shape[1])

    # Create a dictionary that maps feature names to their MI scores
    feature_mi_scores1 = {feature_name: mi_score for feature_name, mi_score in zip(X1.columns, mi_scores1)}

    # Sort the features in ascending order of their MI scores
    sorted_features_ascending1 = sorted(feature_mi_scores1.items(), key=lambda x: x[1])

    # select best 500 and plot MI values

    # Extract the feature names and their MI scores from sorted_features_ascending
    feature_names, mi_scores = zip(*sorted_features_ascending1)

    # Select the last 500 features and their corresponding MI scores
    feature_names = feature_names[-500:]
    mi_scores = mi_scores[-500:]

    # Create a bar plot to visualize MI scores for the last 500 features
    plt.figure(figsize=(12, 8))
    plt.barh(range(len(feature_names)), mi_scores, align='center', alpha=0.7)
    plt.yticks(range(len(feature_names)), feature_names)
    plt.xlabel('Mutual Information Score')
    plt.ylabel('Feature Name')
    plt.title(f'Mutual Information Scores for the Last 500 Features (Ascending Order) Set {i}')
    plt.grid(axis='x', linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

    score_dataset1_export = X1_dataset.copy()
    score_dataset1_export.to_excel(f"partially-processed-score-dataset-set{i}-GSE109068.xlsx", sheet_name=f"GSE109068-score-set{i}")
```
