```python
# Now I will standardize the datasets and then I will concat the datasets
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# df_1_common_features
```

```python
df_1_common_features_X = df_1_common_features.drop(columns=["target_value"])
```

```python
scaler = StandardScaler()
```

```python
df_1_X_scaled_data = scaler.fit_transform(df_1_common_features_X)
```

```python
df_1_X_standardized = pd.DataFrame(df_1_X_scaled_data, columns=df_1_common_features_X.columns)
```

```python
df_1_X_standardized.head()
```

```python
df_1_standardized = df_1_X_standardized.copy()
```

```python
df_1_standardized["target_value"] = df_1_common_features["target_value"]
```

```python
df_1_standardized.head()
```

```python
df_1_standardized
```

```python
# df_2_common_features
```

```python
df_2_common_features_X = df_2_common_features.drop(columns=["target_value"])
```

```python
df_2_X_scaled_data = scaler.fit_transform(df_2_common_features_X)
```

```python
df_2_X_standardized = pd.DataFrame(df_2_X_scaled_data, columns=df_2_common_features_X.columns)
```

```python
df_2_X_standardized.head()
```

```python
df_2_standardized = df_2_X_standardized.copy()
```

```python
df_2_standardized["target_value"] = df_2_common_features["target_value"]
```

```python
df_2_standardized.head()
```

```python
df_2_standardized
```

```python
df_1_standardized.shape
```

```python
df_2_standardized.shape
```

```python
combined_dataset_with_standardization = pd.concat([df_1_standardized, df_2_standardized], ignore_index=True)
```
