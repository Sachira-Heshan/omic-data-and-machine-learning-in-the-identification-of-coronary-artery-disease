```python
# building logistic regression models for different number of features choose by Mutual Information in GSE113079
```

```python
import pandas as pd
```

```python
df = pd.read_excel('/content/drive/MyDrive/Research/Datasets/Cleaned/cleaned-dataset-GSE113079.xlsx')
```

```python
df.head()
```

```python
df = df.T
```

```python
column_names = df.iloc[0, :]
```

```python
df.columns = column_names.tolist()
```

```python
df.head()
```

```python
df.drop(index=["Unnamed: 0"], inplace=True)
```

```python
df.head()
```

```python
# convert target value as integer
df["target_value"] = df["target_value"].astype(int)
```

```python
df.value_counts("target_value")
```

```python
df.shape
```

```python
# divide the features and target values
X = df.drop(columns=["target_value"])
y = df["target_value"]
```

```python
X.shape
```

```python
y.shape
```

```python
# build the logistic regression models
# first try the 100 features selected using mutual information
```

```python
# feature selection using mutual information
import numpy as np
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest

# mutual information selector
mi_classi_100 = SelectKBest(score_func=mutual_info_classif, k=100)

# apply the mutual information selector on the dataset
X_MI_100 = mi_classi_100.fit_transform(X, y)

# get the selected 100 features
selected_features_100 = mi_classi_100.get_support()

# feature names of selected features
selected_feature_names_100 = X.columns[selected_features_100]

X_new_100 = X[selected_feature_names_100]

print("Shape of X_new_100: ", X_new_100.shape)

# divide the dataset to train 0.8 and test 0.2
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new_100, y, test_size=0.2, random_state=42)

# create Logistic Regression classifier and train the data
from sklearn.linear_model import LogisticRegression
logistic_classifier_100 = LogisticRegression(random_state=42)
logistic_classifier_100.fit(X_train, y_train)

# test the model on test set
y_pred_100 = logistic_classifier_100.predict(X_test)
y_pred_100

# evaluate the model using accuracy, classification_report and sensitivity
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

# let's check the accuracy of the model
accuracy_logistic_100_MI = accuracy_score(y_test, y_pred_100)
print(f"Accuracy of the MI 100 Logistic Regression model is: {round(accuracy_logistic_100_MI*100, 3)}%")

# confusion matrix of the model
confusion_matrix_logistic_100_MI = confusion_matrix(y_test, y_pred_100)

TP_logistic_100 = confusion_matrix_logistic_100_MI[1, 1]  # True Positives
FN_logistic_100 = confusion_matrix_logistic_100_MI[1, 0]  # False Negatives

# Calculate Sensitivity (True Positive Rate or Recall)
sensitivity_logistic_100 = TP_logistic_100 / (TP_logistic_100 + FN_logistic_100)
print(f"Sensitivity for 100 features and Logistic Regression model (True Positive Rate): {round(sensitivity_logistic_100*100, 3)}%")

# print the confusion matrix
print("Confusion matrix of the MI 100 Logistic Regression model is: ")
print(confusion_matrix_logistic_100_MI)

# classification report for the
report_logistic_100 = classification_report(y_test, y_pred_100)
print("Classification report of MI 100 Logistic Regression model is: ")
print(report_logistic_100)
```
