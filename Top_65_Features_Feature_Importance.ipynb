```python
import pandas as pd
```

```python
# RNA
df = pd.read_excel('/content/drive/MyDrive/Research/Datasets/Cleaned/cleaned-non-normalized-dataset-GSE208194.xlsx')
df
```

```python
dataset = df.drop(columns=["Unnamed: 0"])
dataset
```

```python
X = dataset.drop(columns=["target_value"])
y = dataset["target_value"]
```

```python
# mutual information values
from sklearn.feature_selection import mutual_info_classif
import numpy as np

# Calculate Mutual Information between each feature and the target variable
mi_scores = mutual_info_classif(X, y, discrete_features=[False]*X.shape[1])

# Create a dictionary that maps feature names to their MI scores
feature_mi_scores = {feature_name: mi_score for feature_name, mi_score in zip(X.columns, mi_scores)}

# Sort the features in ascending order of their MI scores
sorted_features_ascending = sorted(feature_mi_scores.items(), key=lambda x: x[1])
```

```python
# now let's plot this data
import matplotlib.pyplot as plt

# Extract the feature names and their MI scores from sorted_features_ascending
feature_names, mi_scores = zip(*sorted_features_ascending)

# Create a bar plot to visualize MI scores
plt.figure(figsize=(12, 8))
plt.barh(range(len(feature_names)), mi_scores, align='center', alpha=0.7)
plt.yticks(range(len(feature_names)), feature_names)
plt.xlabel('Mutual Information Score')
plt.ylabel('Feature Name')
plt.title('Mutual Information Scores of All Features')
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# prompt: calculate the feature importance values for random forest classifier for above data and plot the data with bigger fonts and clear labels, highest feature importance feature to be in the top of the plot

import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Initialize a Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Fit the classifier to the data
rf_classifier.fit(X, y)

# Get feature importances
feature_importances = rf_classifier.feature_importances_

# Create a dictionary that maps feature names to their importance scores
feature_importance_dict = {feature_name: importance_score for feature_name, importance_score in zip(X.columns, feature_importances)}

# Sort the features in descending order of their importance scores
sorted_features_descending = sorted(feature_importance_dict.items(), key=lambda x: x[1], reverse=True)

# Extract the feature names and their importance scores from sorted_features_descending
feature_names, importance_scores = zip(*sorted_features_descending)

# Create a bar plot to visualize feature importances
plt.figure(figsize=(12, 8))
plt.barh(range(len(feature_names)), importance_scores, align='center', alpha=0.7)
plt.yticks(range(len(feature_names)), feature_names, fontsize=12)  # Increased font size for feature names
plt.xlabel('Feature Importance Score', fontsize=14)  # Increased font size for x-axis label
plt.ylabel('Feature Name', fontsize=14)  # Increased font size for y-axis label
plt.title('Feature Importance Scores from Random Forest', fontsize=16)  # Increased font size for title
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# prompt: plot this for highest 65 features, highest feature importance value feature in the top

import matplotlib.pyplot as plt
# Select the top 65 features
top_65_features = sorted_features_descending[:65]

# Extract the feature names and their importance scores from top_65_features
feature_names_top65, importance_scores_top65 = zip(*top_65_features)

# Create a bar plot to visualize feature importances for the top 65 features
plt.figure(figsize=(12, 12))  # Increased figure height for better visualization
plt.barh(range(len(feature_names_top65)), importance_scores_top65, align='center', alpha=0.7)
plt.yticks(range(len(feature_names_top65)), feature_names_top65, fontsize=10)  # Adjusted font size for feature names
plt.xlabel('Feature Importance Score', fontsize=14)
plt.ylabel('Feature Name', fontsize=14)
plt.title('Top 65 Feature Importance Scores from Random Forest', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# prompt: plot this for highest 25 features, highest feature importance value feature should be in the top

import matplotlib.pyplot as plt
# Select the top 25 features
top_25_features = sorted_features_descending[:25]

# Extract the feature names and their importance scores from top_25_features
feature_names_top25, importance_scores_top25 = zip(*top_25_features)

# Create a bar plot to visualize feature importances for the top 25 features
plt.figure(figsize=(12, 8))
plt.barh(range(len(feature_names_top25)), importance_scores_top25, align='center', alpha=0.7)
plt.yticks(range(len(feature_names_top25)), feature_names_top25, fontsize=10)
plt.xlabel('Feature Importance Score', fontsize=14)
plt.ylabel('Feature Name', fontsize=14)
plt.title('Top 25 Feature Importance Scores from Random Forest', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Invert y-axis to have the highest feature at the top
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```
