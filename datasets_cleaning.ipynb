```python
df = pd.read_excel('/content/drive/MyDrive/Research/Datasets/Partially Processed/partially-processed-dataset-GSE198885.xlsx')
```

```python
df.head()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
# get the column indexes that have more null values
null_counts = df.isnull().sum()
threshold_value = 7
columns_to_drop = null_counts[null_counts >= threshold_value].index
```

```python
columns_to_drop
```

```python
df_drop = df.drop(columns=columns_to_drop)
```

```python
df.shape
```

```python
df_drop.shape
```

```python
df_drop.isnull().sum().sum()
```

```python
df_drop.size
```

```python
# now it is only have almost 1% null values from the whole dataset
# let's fill those null values with suitable values in the dataset
```

```python
df_drop.sample(10)
```

```python
# let's fill the null values with mean value of each column
```

```python
df_fill_with_mean = df_drop.fillna(df_drop.mean())
```

```python
df_fill_with_mean.sample(10)
```

```python
df_fill_with_mean.isnull().sum().sum()
```

```python
# now we don't have any null values in our dataset
```

```python
df_fill_with_mean
```

```python
# GSE198885
df_fill_with_mean.to_excel("cleaned-dataset-GSE198885.xlsx", sheet_name="GSE198885")
```
