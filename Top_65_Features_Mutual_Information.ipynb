```python
import pandas as pd
```

```python
df = pd.read_excel('/content/drive/MyDrive/Research/Datasets/Cleaned/cleaned-non-normalized-dataset-GSE208194.xlsx')
df
```

```python
dataset = df.drop(columns=["Unnamed: 0"])
dataset
```

```python
dataset.value_counts('target_value')
```

```python
dataset.shape
```

```python
X = dataset.drop(columns=["target_value"])
y = dataset["target_value"]
```

```python
# mutual information values
from sklearn.feature_selection import mutual_info_classif
import numpy as np

# Calculate Mutual Information between each feature and the target variable
mi_scores = mutual_info_classif(X, y, discrete_features=[False]*X.shape[1])

# Create a dictionary that maps feature names to their MI scores
feature_mi_scores = {feature_name: mi_score for feature_name, mi_score in zip(X.columns, mi_scores)}

# Sort the features in ascending order of their MI scores
sorted_features_ascending = sorted(feature_mi_scores.items(), key=lambda x: x[1])
```

```python
# now let's plot this data
import matplotlib.pyplot as plt

# Extract the feature names and their MI scores from sorted_features_ascending
feature_names, mi_scores = zip(*sorted_features_ascending)

# Create a bar plot to visualize MI scores
plt.figure(figsize=(12, 8))
plt.barh(range(len(feature_names)), mi_scores, align='center', alpha=0.7)
plt.yticks(range(len(feature_names)), feature_names)
plt.xlabel('Mutual Information Score')
plt.ylabel('Feature Name')
plt.title('Mutual Information Scores of All Features')
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Filter features with non-zero mutual information scores
selected_features = [feature_name for feature_name, mi_score in sorted_features_ascending if mi_score > 0]
# let's create new X with only these features
X_selected = X[selected_features]
X_selected.shape
```

```python
# prompt: need to plot for highest 25 features mutual information values as cell [8] with big font and clear labels

import matplotlib.pyplot as plt
# Sort the features in descending order of their MI scores (for top features)
sorted_features_descending = sorted(feature_mi_scores.items(), key=lambda x: x[1], reverse=True)

# Select the top 25 features
top_25_features = sorted_features_descending[:25]

# Extract the feature names and their MI scores from top_25_features
feature_names_top25, mi_scores_top25 = zip(*top_25_features)

# Create a bar plot to visualize MI scores for the top 25 features
plt.figure(figsize=(12, 8))
plt.barh(range(len(feature_names_top25)), mi_scores_top25, align='center', alpha=0.7)
plt.yticks(range(len(feature_names_top25)), feature_names_top25, fontsize=12)  # Increased font size for y-axis labels
plt.xlabel('Mutual Information Score', fontsize=14)  # Increased font size for x-axis label
plt.ylabel('Feature Name', fontsize=14)  # Increased font size for y-axis label
plt.title('Mutual Information Scores of Top 25 Features', fontsize=16)  # Increased font size for title
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
