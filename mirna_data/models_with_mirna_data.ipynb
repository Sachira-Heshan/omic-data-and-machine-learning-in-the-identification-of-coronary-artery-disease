```python
X = dataset.drop(columns=["target_value"])
y = dataset["target_value"]
```

```python
# mutual information values
from sklearn.feature_selection import mutual_info_classif
import numpy as np

# Calculate Mutual Information between each feature and the target variable
mi_scores = mutual_info_classif(X, y, discrete_features=[False]*X.shape[1])

# Create a dictionary that maps feature names to their MI scores
feature_mi_scores = {feature_name: mi_score for feature_name, mi_score in zip(X.columns, mi_scores)}

# Sort the features in ascending order of their MI scores
sorted_features_ascending = sorted(feature_mi_scores.items(), key=lambda x: x[1])
```

```python
# now let's plot this data
import matplotlib.pyplot as plt

# Extract the feature names and their MI scores from sorted_features_ascending
feature_names, mi_scores = zip(*sorted_features_ascending)

# Create a bar plot to visualize MI scores
plt.figure(figsize=(12, 8))
plt.barh(range(len(feature_names)), mi_scores, align='center', alpha=0.7)
plt.yticks(range(len(feature_names)), feature_names)
plt.xlabel('Mutual Information Score')
plt.ylabel('Feature Name')
plt.title('Mutual Information Scores of All Features')
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Filter features with non-zero mutual information scores
selected_features = [feature_name for feature_name, mi_score in sorted_features_ascending if mi_score > 0]
# let's create new X with only these features
X_selected = X[selected_features]
X_selected.shape
```

```python
# SVM
import numpy as np
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

no_of_features = [1]
cross_val_k = [5, 8, 10]
results = {}

for feature_count in no_of_features:
    print(f"RESULTS FOR {feature_count} FEATURES")
    print("--------------------------------------------------")

    results[feature_count] = {}

    # Feature selection using mutual information
    mi_classif = SelectKBest(score_func=mutual_info_classif, k=feature_count)

    # Apply the mutual information selector on the dataset
    X_MI_selected = mi_classif.fit_transform(X, y)

    # Get the selected features
    selected_features = mi_classif.get_support()

    # Feature names of selected features
    selected_feature_names = X.columns[selected_features]

    X_new = X[selected_feature_names]

    print("Shape of X_new: ", X_new.shape)

    # Divide the dataset to train 0.8 and test 0.2
    X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)

    # Create SVM classifier and fit the data
    svm_classif = SVC()
    svm_classif.fit(X_train, y_train)

    # Test the model on the test set
    y_pred = svm_classif.predict(X_test)

    # Check the accuracy of the model
    accuracy_svm_model = accuracy_score(y_test, y_pred)
    print(f"Accuracy of the {feature_count} features SVM model is: {round(accuracy_svm_model * 100, 3)}%")

    results[feature_count]["accuracy"] = round(accuracy_svm_model * 100, 3)

    # Confusion matrix of the model
    confusion_matrix_svm_model = confusion_matrix(y_test, y_pred)

    TP_svm_model = confusion_matrix_svm_model[1, 1]  # True Positives
    FN_svm_model = confusion_matrix_svm_model[1, 0]  # False Negatives

    # Calculate Sensitivity (True Positive Rate or Recall)
    sensitivity_svm_model = TP_svm_model / (TP_svm_model + FN_svm_model)
    print(f"Sensitivity for {feature_count} features SVM model (True Positive Rate): {round(sensitivity_svm_model * 100, 3)}%")

    results[feature_count]["sensitivity"] = round(sensitivity_svm_model * 100, 3)

    # Print the confusion matrix
    print(f"Confusion matrix of the {feature_count} features SVM model is: ")
    print(confusion_matrix_svm_model)

    # Classification report for the model
    report_svm_model = classification_report(y_test, y_pred)
    print(f"Classification report of {feature_count} features SVM model is: ")
    print(report_svm_model)

    # Now let's do the cross-validation
    for k in cross_val_k:
        print(f"{k}-fold Cross Validation")

        results[feature_count][f"{k}-fold-CV"] = {}

        # Create a cross-validation object (K-Fold cross-validator)
        kf = KFold(n_splits=k, shuffle=True, random_state=42)

        # Perform k-fold cross-validation and get the cross-validation scores using SVM
        scores = cross_val_score(svm_classif, X_new, y, cv=kf, scoring='accuracy')

        # Print the cross-validation scores
        print("Cross-Validation Scores:", scores)

        # Calculate and print the mean and standard deviation of the scores
        print(f"Mean Accuracy: {round(scores.mean() * 100, 3)}%")
        print(f"Standard Deviation: {round(scores.std() * 100, 3)}%")

        results[feature_count][f"{k}-fold-CV"]["mean-accuracy"] = round(scores.mean() * 100, 3)
        results[feature_count][f"{k}-fold-CV"]["standard-deviation"] = round(scores.std() * 100, 3)
        results[feature_count]["acc-std"] = f"{results[feature_count][f'{k}-fold-CV']['mean-accuracy']}% ± {results[feature_count][f'{k}-fold-CV']['standard-deviation']}%"

        print("")
    print("")
```

```python
# Logistic Regression
import numpy as np
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

no_of_features = [5, 15, 25, 35, 45, 55, 65]
cross_val_k = [5, 8, 10]
results_lr = {}

for feature_count in no_of_features:
    print(f"RESULTS FOR {feature_count} FEATURES")
    print("--------------------------------------------------")

    results_lr[feature_count] = {}

    # Feature selection using mutual information
    mi_classif = SelectKBest(score_func=mutual_info_classif, k=feature_count)

    # Apply the mutual information selector on the dataset
    X_MI_selected = mi_classif.fit_transform(X, y)

    # Get the selected features
    selected_features = mi_classif.get_support()

    # Feature names of selected features
    selected_feature_names = X.columns[selected_features]

    X_new = X[selected_feature_names]

    print("Shape of X_new: ", X_new.shape)

    # Divide the dataset to train 0.8 and test 0.2
    X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)

    # Create Logistic Regression classifier and fit the data
    lr_classif = LogisticRegression()
    lr_classif.fit(X_train, y_train)

    # Test the model on the test set
    y_pred = lr_classif.predict(X_test)

    # Check the accuracy of the model
    accuracy_lr_model = accuracy_score(y_test, y_pred)
    print(f"Accuracy of the {feature_count} features Logistic Regression model is: {round(accuracy_lr_model * 100, 3)}%")

    results_lr[feature_count]["accuracy"] = round(accuracy_lr_model * 100, 3)

    # Confusion matrix of the model
    confusion_matrix_lr_model = confusion_matrix(y_test, y_pred)

    TP_lr_model = confusion_matrix_lr_model[1, 1]  # True Positives
    FN_lr_model = confusion_matrix_lr_model[1, 0]  # False Negatives

    # Calculate Sensitivity (True Positive Rate or Recall)
    sensitivity_lr_model = TP_lr_model / (TP_lr_model + FN_lr_model)
    print(f"Sensitivity for {feature_count} features Logistic Regression model (True Positive Rate): {round(sensitivity_lr_model * 100, 3)}%")

    results_lr[feature_count]["sensitivity"] = round(sensitivity_lr_model * 100, 3)

    # Print the confusion matrix
    print(f"Confusion matrix of the {feature_count} features Logistic Regression model is: ")
    print(confusion_matrix_lr_model)

    # Classification report for the model
    report_lr_model = classification_report(y_test, y_pred)
    print(f"Classification report of {feature_count} features Logistic Regression model is: ")
    print(report_lr_model)

    # Now let's do the cross-validation
    for k in cross_val_k:
        print(f"{k}-fold Cross Validation")

        results_lr[feature_count][f"{k}-fold-CV"] = {}

        # Create a cross-validation object (K-Fold cross-validator)
        kf = KFold(n_splits=k, shuffle=True, random_state=42)

        # Perform k-fold cross-validation and get the cross-validation scores using Logistic Regression
        scores = cross_val_score(lr_classif, X_new, y, cv=kf, scoring='accuracy')

        # Print the cross-validation scores
        print("Cross-Validation Scores:", scores)

        # Calculate and print the mean and standard deviation of the scores
        print(f"Mean Accuracy: {round(scores.mean() * 100, 3)}%")
        print(f"Standard Deviation: {round(scores.std() * 100, 3)}%")

        results_lr[feature_count][f"{k}-fold-CV"]["mean-accuracy"] = round(scores.mean() * 100, 3)
        results_lr[feature_count][f"{k}-fold-CV"]["standard-deviation"] = round(scores.std() * 100, 3)
        results_lr[feature_count]["acc-std"] = f"{results_lr[feature_count][f'{k}-fold-CV']['mean-accuracy']}% ± {results_lr[feature_count][f'{k}-fold-CV']['standard-deviation']}%"

        print("")
    print("")
```
