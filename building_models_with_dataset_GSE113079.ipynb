```python
import pandas as pd
```

```python
df = pd.read_excel('/content/drive/MyDrive/Research/Datasets/Cleaned/cleaned-dataset-GSE113079.xlsx')
```

```python
df = df.T
```

```python
column_names = df.iloc[0, :]
```

```python
df.columns = column_names.tolist()
```

```python
df.head()
```

```python
df.drop(index=["Unnamed: 0"], inplace=True)
```

```python
df.head()
```

```python
# convert target value as integer
df["target_value"] = df["target_value"].astype(int)
```

```python
df.head()
```

```python
df.shape
```

```python
# data is already standardized and no null values
```

```python
# now let's apply some feature selection techniques to select best features to build the models
```

```python
# feature selection using mutual information
import numpy as np
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
```

```python
# mutual information classifier
mi_classi = SelectKBest(score_func=mutual_info_classif, k=125)
```

```python
# divide the features and target values
X = df.drop(columns=["target_value"])
y = df["target_value"]
```

```python
X.head()
```

```python
# now apply the mutual information on the features X
```

```python
# apply the mutual information classifier on the dataset
X_MI = mi_classi.fit_transform(X, y)
```

```python
selected_features = mi_classi.get_support()
```

```python
selected_feature_names = X.columns[selected_features]
```

```python
X_MI.shape
```

```python
X_new = X[selected_feature_names]
```

```python
X_new.head()
```

```python
X_new.shape
```

```python
# divide the dataset to train 0.8 and test 0.2
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
# create the dataset with selected features and target values
MI_selected_df = X_new.copy()
MI_selected_df["target_value"] = y
MI_selected_df
```

```python
# apply SVM algorithm on the data and build a model with 125 features
from sklearn.svm import SVC
svm_classif = SVC()
svm_classif.fit(X_train, y_train)
```

```python
# test the model on test set
pred = svm_classif.predict(X_test)
```

```python
pred
```

```python
# now let's evaluate the built model using some evaluation metrics
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# let's check the accuracy of the model
accuracy_svm_125_MI = accuracy_score(pred, y_test)
print("Accuracy of the MI 125 SVM model is: ", accuracy_svm_125_MI)
```

```python
# confusion matrix of the model
confusion_matrix_svm_125_MI = confusion_matrix(pred, y_test)
print("Confusion matrix of the MI 125 SVM model is: ", confusion_matrix_svm_125_MI)
```

```python
TP = confusion_matrix_svm_125_MI[1, 1]  # True Positives
FN = confusion_matrix_svm_125_MI[1, 0]  # False Negatives

# Calculate Sensitivity (True Positive Rate or Recall)
sensitivity = TP / (TP + FN)

print(f"Sensitivity (True Positive Rate): {sensitivity}")
```
