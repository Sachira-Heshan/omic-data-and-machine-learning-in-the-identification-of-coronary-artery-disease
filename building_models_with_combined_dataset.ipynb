```python
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/Research/Datasets/Cleaned/combined-standardized-datasets-GSE230165-n-GSE198885.xlsx")
```

```python
df.head()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df
```

```python
df.value_counts("target_value")
```

```python
df.isnull().sum().sum()
```

```python
# seperate features and targets
X = df.drop(columns=["target_value"])
y = df["target_value"]
```

```python
X.head()
```

```python
y.tolist()
```

```python
# building model with algorithm SVM and feature selection using mutual information
```

```python
# feature selection using mutual information
import numpy as np
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
# mutual information selector
mi_classi_75 = SelectKBest(score_func=mutual_info_classif, k=75)
# apply the mutual information selector on the dataset
X_MI_75 = mi_classi_75.fit_transform(X, y)
# get the selected 75 features
selected_features_75 = mi_classi_75.get_support()

# feature names of selected features
selected_feature_names_75 = X.columns[selected_features_75]
X_MI_75.shape
X_new_75 = X[selected_feature_names_75]
X_new_75.head()

# divide the dataset to train 0.8 and test 0.2
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new_75, y, test_size=0.2, random_state=42)

X_train.shape

# create the dataset with selected features and target values
MI_selected_df_75 = X_new_75.copy()
MI_selected_df_75["target_value"] = y
MI_selected_df_75

# create SVM classifier and fit the data
from sklearn.svm import SVC
svm_classif_75 = SVC()
svm_classif_75.fit(X_train, y_train)

# test the model on test set
pred_75 = svm_classif_75.predict(X_test)
pred_75

# evaluate the model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
# let's check the accuracy of the model
accuracy_svm_75_MI = accuracy_score(pred_75, y_test)
print("Accuracy of the MI 100 SVM model is: ", accuracy_svm_75_MI)

# confusion matrix of the model
confusion_matrix_svm_75_MI = confusion_matrix(pred_75, y_test)
print("Confusion matrix of the MI 75 SVM model is: ", confusion_matrix_svm_75_MI)

TP_75 = confusion_matrix_svm_75_MI[1, 1]  # True Positives
FN_75 = confusion_matrix_svm_75_MI[1, 0]  # False Negatives

# Calculate Sensitivity (True Positive Rate or Recall)
sensitivity_75 = TP_75 / (TP_75 + FN_75)

print(f"Sensitivity for 75 features (True Positive Rate): {sensitivity_75}")
```
